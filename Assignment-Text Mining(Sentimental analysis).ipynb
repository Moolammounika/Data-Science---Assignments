```python
from nltk import word_tokenize

def stopwords(var):
    tokens=word_tokenize(var)
    return[x for x in tokens if x not in stop_words]       
```

```python
import nltk
nltk.download('punkt')
```

```python
#for positive score
def positive_score(var):
    p=[]
    tokens=word_tokenize(var)
    for i in tokens:
        if i in positive:
            p.append(i)
    return len(p)
```

```python
data['postive_score']=data['clean1'].apply(positive_score)
```

```python
#for negative score
def negative_score(var):
    n=[]
    tokens=word_tokenize(var)
    for i in tokens:
        if i in negative:
            n.append(i)
    return len(n)
```

```python
data['negative_score']=data['clean1'].apply(negative_score)
```

```python
def tot_words(var):
    p=[]
    tokens=word_tokenize(var)
    for i in tokens:
        if i not in stop_words:
            p.append(i)
    return len(p)
```

```python
data['Total_words']=data['clean1'].apply(tot_words)
```
