### Book Dataset

```python
#importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
# reading the data
data=pd.read_csv('book (4).csv',encoding="latin~1")
data
```

```python
data[0:5]
```

```python
#to know the unique values in that column in a numeric value
data["Book.Title"].nunique()
```

```python
#to know the unique values in that column
data["Book.Title"].unique()
```

```python
df=data.iloc[:,1:]
df.head()
```

```python
#to know duplicated values
df["Book.Rating"].isna().sum()
```

```python
#to know unique values in that column
data["User.ID"].nunique()
```

```python
#shape of the data
df.shape
```

```python
#pivot table
df1=df.pivot_table(index='User.ID',
                     columns='Book.Title',
                     values='Book.Rating').reset_index(drop=True)
df1
```

```python
#filling null value with zeroes
df1.fillna(0,inplace=True)
```

```python
df1
```

### Cosine similarities

```python
df2= 1-pairwise_distances(df1,metric='cosine')
df2
```

```python
df3=pd.DataFrame(df2)
df3.head()
```

```python
#setting the index and column names with User.ID
df3.index=df['User.ID'].unique()
df3.columns=df['User.ID'].unique()
```

```python
df3
```

```python
np.fill_diagonal(df2,0)
df3.iloc[0:5,0:5]
```

```python
# most similar users
df3.idxmax(axis=1)
```
