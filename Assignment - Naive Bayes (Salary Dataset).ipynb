### Gaussian model

```python
#building a model
from sklearn.naive_bayes import GaussianNB as GB
model1=GB()
model1.fit(x,y)
```

```python
#predictng the model
pred1=model1.predict(x1)
```

```python
pred1
```

```python
#testing accuacy
accuracy = np.mean(pred1==y1)
```

### Multinomial model

```python
from sklearn.naive_bayes import MultinomialNB as MB
```

```python
classifier_mb = MB()
classifier_mb.fit(x,y)
```

```python
#predicting the model
pred1 = classifier_mb.predict(x1)
```
