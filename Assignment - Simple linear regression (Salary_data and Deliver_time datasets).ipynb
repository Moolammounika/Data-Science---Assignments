### salary_data dataset

```python
#importing the libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#reading the data
data=pd.read_csv('Salary_Data.csv')
data.head()
```

```python
#checking for null values
data.isnull()
```

```python
x=data[['YearsExperience']]
x
```

```python
y=data[['Salary']]
y
```

```python
#plotting the data
plt.scatter(x,y,color='blue',marker='.')
plt
```

```python
#Linear regression and fit the model
classifier = LinearRegression()
classifier.fit(x,y)
```

```python
#Predict for X dataset
y_pred = classifier.predict(x)
y_pred
```

```python
classifier.coef_
```

```python
classifier.intercept_
```

```python
9449.96232146*10.5+25792.20019867
```

```python
#predicting the model
data['prediction']=y_pred
```

```python
data
```

```python
classifier.score(x,y)
```

```python
#plotting a graph
plt.xlabel('Years of Experience', fontsize=15)
plt.ylabel('salary' ,fontsize=15)
plt.scatter(data.YearsExperience,data.Salary,color='blue',marker='.')
plt.plot(data.YearsExperience,y_pred,color='pink')
```

```python
from sklearn.metrics import r2_score
sol=r2_score(y,y_pred)
```

### Deliver time

```python
#importing the librares
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading the data
data = pd.read_csv('delivery_time.csv')
data.head(10)
```

```python
df= pd.DataFrame(data)
```

```python
df
```

```python
#checking for null values
df.isna()
```

```python
#calculating satistical values
df.describe()
```

```python
#plottng the data
sns.distplot(df['Delivery Time'])
```

```python
sns.distplot(df['Sorting Time'])
```

```python
plt.plot(df['Delivery Time'],df['Sorting Time'],"bo")
plt.xlabel("Delivery Time")
plt.ylabel("Sorting Time")
```

```python
#Correlation
data.corr()
```

```python
#rename dataset
data= data.rename(columns = {'Delivery Time': 'DT', 'Sorting Time': 'ST'}, inplace = False)
```

```python
# Ordinary Least Squares(OLS) 
import statsmodels.formula.api as smf
model = smf.ols("DT~ST",data = data).fit()
```

```python
data
```

```python
#Coefficients
model.params
```

```python
# Regression plot
sns.regplot(x="ST", y="DT", data=data)
```

### t and p-Values
print(model.tvalues, '\n', model.pvalues)

### P-value of ST is significant because it's is less than 0.05.

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

### R-Square value is 68% . which is not good output. So do the ols again to get better output.

```python
model.summary()
```

```python
# Distplot for both DT and ST

data_1=data
data_1['DT'] = np.log(data_1['DT'])
data_1['ST'] = np.log(data_1['ST'])
sns.distplot(data_1['DT'])
fig = plt.figure()
sns.distplot(data_1['ST'])
fig = plt.figure()
```

### model2

```python
model_2 = smf.ols("DT~np.log(ST)",data=data).fit()
```

```python
model_2
```

```python
model_2.summary()
```

### Hence the P-value is less than 0.05. So X(log(ST)) varibale is significance and also R-Square value is 0.77 That’s mean this model will predict the output 77% time correct. So we need still better model so do again ols.

```python
# Distplot for both DT abd ST

data_2=data
data_1['DT'] = np.log(data_1['DT'])
data_1['ST'] = np.log(data_1['ST'])
sns.distplot(data_1['DT'])
fig = plt.figure()
sns.distplot(data_1['ST'])
fig = plt.figure()
```

```python
model_3 = smf.ols("np.log(DT)~ST",data=data).fit()
```

```python
model_3.summary()
```

### Hence the P-value is less than 0.05. So X(ST) varibale is significance and also R-Square value is 0.82 That’s mean this model will predict the output 82% time correct.

```python
# Distplot for both DT abd ST

data_3=data
data_1['DT'] = np.log(data_1['DT'])
data_1['ST'] = np.log(data_1['ST'])
sns.distplot(data_1['DT'])
fig = plt.figure()
sns.distplot(data_1['ST'])
fig = plt.figure()
```

### R-Squared value of the given Delivery_time_dataset model3 output is obtained as 82%. which means it's good model.
