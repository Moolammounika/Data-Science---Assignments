```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
import tensorflow
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(11,kernel_initializer='random_normal',activation='relu'))
model.add(Dense(8,kernel_initializer='random_normal',activation='linear'))
model.add(Dense(1,kernel_initializer='random_normal',activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(x,y,validation_split=0.03, epochs=150, batch_size=10)
```

```python
score = model.evaluate(x,y)
```

```python
print("%s: %.2f%%" % (model.metrics_names[1], score[1]*100))
```

```python
model.history.history.keys()
```

```python
import matplotlib.pyplot as plt
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='lower right')
plt.show()

```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss') 
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
